<h4 style="color: black;">TYPES OF STREAMERS</h4>

```python
df['TYPE'] = np.where(df['TYPE'] == 'personality', 'Casual', 'esports')
```

```python
type_streamers = df['TYPE'].value_counts()
type_streamers
```

<h4 style="color: black;">POPULAR LANGUAGES ON TWITCH</h4>

```python
language_streamers = df.groupby('LANGUAGE')['NAME'].count().sort_values(ascending=False)[:5]
plt.figure(figsize=(10, 6))
language_streamers.plot(kind='bar')
plt.title("TOP 5 LANGUAGES STREAMED IN")
plt.xticks(rotation = 45)
plt.xlabel("LANGUAGES")
plt.ylabel("STREAMER COUNT")
plt.grid()
plt.show()
```

<h4 style="color: black;">MOST POPULAR GAMES</h4>

```python
second_popular_games = df['2ND_MOST_STREAMED_GAME'].value_counts()[:5]
second_popular_games
```

<h4 style="color: black;">MOST FOLLOWED</h4>

```python
sorted_followers = df.sort_values(by="TOTAL_FOLLOWERS", ascending=False)
sorted_followers[['NAME', 'LANGUAGE', 'TOTAL_FOLLOWERS']][:10]
```

<h4 style="color: black;">HIGHEST TOTAL VIEWS</h4>

```python
sorted_total_views = df.sort_values(by="TOTAL_VIEWS", ascending=False)
sorted_total_views[['NAME', 'LANGUAGE', 'TOTAL_VIEWS']][:10]
```

<h4 style="color: black;">HIGHEST VIEWS PER STREAM</h4>

```python
sorted_viewers_per_stream = df.sort_values(by="AVG_VIEWERS_PER_STREAM", ascending=False)
sorted_viewers_per_stream[['NAME', 'LANGUAGE', 'AVG_VIEWERS_PER_STREAM']][:10]
```

<h4 style="color: black;">HIGHEST AVG STREAM TIME</h4>

```python
sorted_avg_stream = df.sort_values(by="AVERAGE_STREAM_DURATION", ascending=False)
sorted_avg_stream[['NAME', 'LANGUAGE', 'AVERAGE_STREAM_DURATION']][:10]
```

<h4 style="color: black;">MOST FOLLOWERS GAINED PER STREAM</h4>

```python
sorted_followers_per_stream = df.sort_values(by="FOLLOWERS_GAINED_PER_STREAM", ascending=False)
sorted_followers_per_stream[['NAME', 'LANGUAGE', 'FOLLOWERS_GAINED_PER_STREAM']][:10]
```

<h4 style="color: black;">WEEKLY ACTIVITY ON TWITCH</h4>

```python
active_days = df['MOST_ACTIVE_DAY'].value_counts()
act_ax = active_days.plot(kind='pie',
                 autopct= "%1.1f%%",
                 title="Most Active Days",
                 wedgeprops=dict(edgecolor='black', linewidth=1),
                 startangle = 140)
plt.figure(figsize=(14,8))
act_ax.set_ylabel(" ")
plt.show()
```

<h4 style="color: black;">VIEWS-FOLLOWERS RATIO</h4>

```python
views_followers_ratio = df[["NAME", "LANGUAGE", "TYPE", "AVG_VIEWERS_PER_STREAM", "TOTAL_FOLLOWERS"]]
views_followers_ratio['VIEWS_FOLLOWERS_RATIO(%)'] = (views_followers_ratio['AVG_VIEWERS_PER_STREAM'] / views_followers_ratio['TOTAL_FOLLOWERS']) * 100
```

```python
views_followers_ratio
```

```python
views_followers_ratio.sort_values(by='VIEWS_FOLLOWERS_RATIO(%)', ascending=False)[:10]
```
